```python
import yfinance as yf
def get_data(symbol: str):
    data = yf.download(tickers='BTC-USD', period='max', interval='1d')
    data = data.xs("BTC-USD", axis=1, level="Ticker")
    data.reset_index(inplace=True, drop=True)
    return data
# Get the data
data = get_data('BTC-USD')
```

```python
import pandas_ta as ta

# function to calculate simple mean average (SMA)
def calculate_sma(data, length: int):
    return ta.sma(data['Close'], length)


data['SMA'] = calculate_sma(data, 20)
data.dropna(inplace=True)
data.reset_index(inplace=True, drop=True)
```

# 1 - Slope of MAs

```python
import numpy as np

def calculate_slope(series, period: int=5):
    slopes = [0 for _ in range(period-1)]
    for i in range(period-1, len(series)):
        x = np.arange(period)
        y = series[i-period+1:i+1].values
        coeff = np.polyfit(x,y,1) # Perform linear fit
        slope = coeff[0]
        percentage_slope = (slope/y[0])*100 # Convert the slope to a percentage
        slopes.append(percentage_slope) # High +ve strong upward trend High -ve strong downward trend
    
    return slopes
```

```python
data['Slope'] = calculate_slope(data['SMA'])
```
